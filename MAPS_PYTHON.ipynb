```python
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx

# Caminho para o shapefile da Baía de Guanabara
shapefile_path = 'caminho/para/o/shapefile/guanabara_bay.shp'

# Ler o shapefile usando geopandas
gdf = gpd.read_file(shapefile_path)

# Criar uma figura e eixos
fig, ax = plt.subplots(figsize=(10, 10))

# Plotar o mapa da Baía de Guanabara
gdf.plot(ax=ax, color='lightblue', edgecolor='black')

# Adicionar um fundo de mapa (tiles) usando contextily
ctx.add_basemap(ax, crs=gdf.crs.to_string(), source=ctx.providers.Stamen.Terrain)

# Configurar o título e os eixos
ax.set_title('Mapa da Baía de Guanabara', fontsize=15)
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Mostrar o mapa
plt.show()
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import matplotlib.patches as mpatches
import numpy as np

# Crie uma projeção da Baía de Guanabara
projecao = ccrs.PlateCarree()
extent = [-43.5, -42.5, -23.1, -22.6]  # longitude e latitude da Baía de Guanabara

# Crie o plot
fig, ax = plt.subplots(subplot_kw={'projection': projecao})
ax.set_extent(extent)
ax.coastlines(resolution='10m')
ax.gridlines(draw_labels=True)

# Adicione título
plt.title('Baía de Guanabara - Rio de Janeiro')

# Coordenadas originais da mancha de óleo (2km²)
original_oil_spill_lon = [-43.25, -43.2, -43.15, -43.2, -43.25]
original_oil_spill_lat = [-22.8, -22.75, -22.8, -22.85, -22.8]

# Calcule o deslocamento após 2 horas (10KT, 110º)
# Velocidade em nós
speed_knots = 10
# Direção em graus
direction_degrees = 110

# Converta a direção de graus para radianos
direction_radians = np.deg2rad(direction_degrees)

# Calcule o deslocamento em coordenadas de latitude e longitude
# Distância em graus (1 grau de latitude ~ 111km, 1 grau de longitude ~ 111km * cos(latitude))
distance_degrees = (speed_knots * 1.852) / (111 * np.cos(np.deg2rad((original_oil_spill_lat[0] + original_oil_spill_lat[2]) / 2)))
# Deslocamento em longitude e latitude
delta_lon = distance_degrees * np.sin(direction_radians)
delta_lat = distance_degrees * np.cos(direction_radians)

# Calcule as novas coordenadas da mancha de óleo após 2 horas
new_oil_spill_lon = [lon + 2 * delta_lon for lon in original_oil_spill_lon]
new_oil_spill_lat = [lat + 2 * delta_lat for lat in original_oil_spill_lat]

# Adicione o polígono da mancha de óleo com as novas coordenadas ao plot
ax.fill(new_oil_spill_lon, new_oil_spill_lat, color='black', transform=projecao)

# Adicione uma legenda para a mancha de óleo
oil_legend = mpatches.Patch(color='black', label='Mancha de óleo (2km²)')
plt.legend(handles=[oil_legend])

plt.show()
```
